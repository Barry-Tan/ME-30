```python
def ave_price():
    fin = open ('prices.dat.txt','r')
    prices = fin.readlines()
    sum_of_price = 0
    number_of_price = 0
    for price in prices:
        to_int = int(price)
        sum_of_price+=to_int
        number_of_price+=1
    ave=sum_of_price/number_of_price
    
    return(ave)
ave=ave_price()

def stand_d(ave):
    fin = open ('prices.dat.txt','r')
    prices = fin.readlines()
    stand_2 = 0
    num_of_price = 0
    print("ave(): " + str(ave))
    print("ave: " + str(ave))      
    for price in prices:
        to_int = int(price)
        stand_1=(to_int-ave)**2
        stand_2+=stand_1
        num_of_price+=1
    stand_3=stand_2/(num_of_price-1)
    print('Stand derviation is:')
    return(stand_3**(0.5))
    """stand derviation = squrt{sigma[(x_i-x_ave)^2/(n-1)]}
       stand_1 means (x_i-x_ave)^2
       stand_2 means summation or sigma of stand_1
       stand_3 means stand_2/(n-1)"""
std=stand_d

def store_file():
    fout = open('price_summary.dat.txt','w')
    fout.write(str(ave)+ 'is ave'+str(std)+ 'is stand derivation')
    fout.close()
    
```
