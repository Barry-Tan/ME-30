```python
import csv
import numpy as np
from statistics import mean
import matplotlib.pyplot as pl
from matplotlib import style

common = []  #common is list of years that all CSVs cover in the form of list of the years in string form.
def shareyears(longtimeline, a, b, c):  #will need to be revise for fourth CSV timeline
    for x in longtimeline:
        if x in a and x in b and x in c:
            common.append(x)
   
def commoncvt(shareyr, shareyrint):  #takes years in list of common years and converts them into integers for matplotlib
    for years in shareyr:
        shareyrint.append(int(years))
    return shareyrint

def aggreg(csvlist, shareline):
    monthlist = []
    yr_avg = []
    for item in shareline[0:-2]:
        monthly = (csvlist[item])
        monthlist.append(monthly)
    for gamma in monthlist:
        for zulu in range(12):
            gamma[zulu] = float(gamma[zulu])
        yr_avg.append(sum(gamma))
    return yr_avg

with open("Average_Wage_Index_1973-2017.csv") as f:
    firstColumn = [line.split(',')[0] for line in f] #### make a list of the years going down the column in csv file. .split used for to seperate strings with comma and also only specify 0th term 
    del firstColumn[0]
li1 = firstColumn

with open("Average_Wage_Index_1973-2017.csv") as f:
    secondColumn = [line.split()[1] for line in f]
    del secondColumn[0]

li7 = []
listModify(li7)
li8 = [] #### list for Wages
for x in range(len(li7)):
    li8.append((float(li7[x]))) #### converts elements of the wage list to floats


grad = open("College_1971-2018.csv", "r+", encoding="ISO-8859-1") 
readerz = csv.reader(grad)
dict1 = {}
dict2 = {}
dict3 = {}
collo1 = [] #### empty list for Private Nonprofit Four-Year 
collo2 = [] #### empty list for Public Four-Year
collo3 = [] #### empty list for Public Two-Year
for job in readerz:
    dict1[job[1]] = [(job[0])] #### assign 1st list to 0th list from csv file
    dict2[job[3]] = [(job[0])] #### same 
    dict3[job[5]] = [(job[0])] #### same
    collo1.append(job[1])
    collo2.append(job[3])
    collo3.append(job[5])
collo1 = [x for x in collo1 if x != collo1[0]] #### removes first element from the list the college list which is a string with open quotations
collo2 = [x for x in collo2 if x != collo2[0]] #### same thing
collo3 = [x for x in collo3 if x != collo3[0]] #### same thing

print(" ")

def listModify2(TL): #### function created for college list for the following reasons
    GL = []
    GP = []
    for k in range(3,49):
        lisp3 = TL[k].replace('$','') #### removes dollar signs from strings in college list
        GL.append(lisp3)
    for p in range(len(GL)):
        lisp4 = GL[p].replace(",","") #### removes commas from strings in the college list
        GP.append(lisp4)
    return GP
colly1 = listModify2(collo1)
colly2 = listModify2(collo2)
colly3 = listModify2(collo3)
def listModify(FL):
    for i in range(len(secondColumn)):
        list2 = secondColumn[i].replace(',','') #### create new value that will replace all commas
        FL.append(list2) #### apply changes to a new list
    return FL

def percentIncrease(g):
    pct = []
    for i in range(len(g[0:-1])): #### consider range of index; cannot calc change of last entry to an index that does not exist
        list4 = (g[i+1]-g[i]) / g[i+1] * 100
        pct.append(list4)
    return pct

def annualAvgIncrease(yrlist):
    avgyr = []
    for i in range(len(yrlist[0:-1])):
        list5 = (yrlist[i+1]-yrlist[i])
        avgyr.append(list5)
    return mean(avgyr)

def maxlen(first, second, third): #### function that determines the max length of each list 
    if len(first) >= len(second) and len(third):
        return first
    elif len(second) >= len (first) and len(third):
        return second
    else:
        return third
def conversion(KL): #### function that takes in any list and converts the values into floats
    college = []
    for x in range(len(KL)):
        college.append((float(KL[x]))) 
    return college
collee1 = conversion(colly1) #### 3 lists that have commas and $ removed and all converted into strings
collee2 = conversion(colly2)
collee3 = conversion(colly3)
tuit = []
for cost in range(len(collee1)): 
    tuit.append(mean([ collee1[cost], collee2[cost], collee3[cost] ])) #### takes the mean of every element and attaches to an empty list (for example li1= [2,3,4] and li2 = [7,4,2] so li3 = [9/2, 7/2,4] )

f = open("Fresh_vegetables_1947-2015.csv", "r+", encoding="ISO-8859-1")
reader = csv.reader(f)
vegcsv = {}                  
vegyr = []    #list of years covered by vegetable csv as strings
for row in reader:
    vegcsv[row[0]] = ([(row[1]), (row[2]),(row[3]),(row[4]),(row[5]),(row[6]),(row[7]),\
                         (row[8]),(row[9]),(row[10]),(row[11]),(row[12])])
    vegyr.append(row[0])
    
    
cow = open("Slaughter_steers_and_heifers_1947-2015.csv", "r+", encoding="ISO-8859-1") #### Open file
readerp = csv.reader(cow)
steercsv = {}              
slag = []             #list of years covered by cow csv as strings
for bro in readerp: #### create a list from csv file
    
    steercsv[bro[0]] = ([(bro[1]), (bro[2]),(bro[3]),(bro[4]),(bro[5]),(bro[6]),(bro[7]),\
                         (bro[8]),(bro[9]),(bro[10]),(bro[11]),(bro[12])])
    slag.append(bro[0])



top_dog = maxlen(li1, vegyr, slag)
shareyears(top_dog, li1, vegyr, slag)

vegtemp = []
steertemp = []
edutemp = []
wagetemp = []

veg_annual = aggreg(vegcsv, common)
slag_annual = aggreg(steercsv, common)


prob2 = "Average Annual Increase (USD)"
prob3 = "Average Annual Amount (USD)"


wage1 = mean(li8)
wage3 = annualAvgIncrease(li8)
wage4 = percentIncrease(li8)

veg1 = mean(veg_annual)
veg3 = annualAvgIncrease(veg_annual)
veg4 = percentIncrease(veg_annual)

slag1 = mean(slag_annual)
slag3 = annualAvgIncrease(slag_annual)
slag4 = percentIncrease(slag_annual)

college1 = mean(tuit)
college3 = annualAvgIncrease(tuit)
college4 = percentIncrease(tuit)

commonint = []  

commoncvt(common, commonint)

def scalar(new, orig):
    for kilo in orig:
        new.append(kilo*100)
    
veg_graph = []

scalar(veg_graph, veg_annual)
       

def thepackage():
    try:
        print(' {0:^20s} | {1:^30s} | {2:^30s}'.format('', prob2, prob3))
        print('|{0:^20s} | {1:^30.2f} | {2:^30.2f}|'.format("Wages", wage3, wage1))
        print('|{0:^20s} | {1:^30.2f} | {2:^30.2f}|'.format("Vegetables", veg3, veg1))
        print('|{0:^20s} | {1:^30.2f} | {2:^30.2f}|'.format("Steers & Heifers", slag3, slag1))
        print('|{0:^20s} | {1:^30.2f} | {2:^30.2f}|'.format("Tuition and Fees*", college3, college1))
        print('')
        print('*Per academic year of higher education')
        print('')
        print("Annual Percent Change of Quantity Per Year:")
        print('')
        print('|{0:^20s} | {1:^20s} | {2:^20s} | {3:^20s} | {4:^20s}'.format("Year", "Wage/Salary", "Cost of Vegetables", "Tuition and Fees", "Steer & Heifer Slaughter"))
        for perc in range(len(common)):
            print('|{0:^20s} | {1:^20.2f} | {2:^20.2f} | {3:^20.2f} | {4:^20.2f}'.format(common[perc], wage4[perc], veg4[perc], college4[perc], slag4[perc]))
    except:
        pass
    
print(thepackage())
    
def sendit(name):
    try:
        with open(name, 'w') as final:
            for i in range(len(veg4)):
                final.write('|{0:^20s} | {1:^30s} | {2:^30s} %d\r\n'.format('', prob2, prob3)% (i+1))
                final.write('|{0:^20s} | {1:^30.2f} | {2:^30.2f}| %d\r\n'.format("Wages", wage3, wage1)% (i+1))
                final.write('|{0:^20s} | {1:^30.2f} | {2:^30.2f}| %d\r\n'.format("Vegetables", veg3, veg1)% (i+1))
                final.write('|{0:^20s} | {1:^30.2f} | {2:^30.2f}| %d\r\n'.format("Steers & Heifers", slag3, slag1)% (i+1))
                final.write('|{0:^20s} | {1:^30.2f} | {2:^30.2f}| %d\r\n'.format("Tuition and Fees", college3, college1)% (i+1))
                final.write('%d\r\n' %(i+1))
                final.write('* Per academic year of higher education %d\r\n' %(i+1))
                final.write('%d\r\n' %(i+1))
                final.write("Annual Percent Change of Quantity Cost Per Year:%d\r\n"% (i+1))
                final.write('%d\r\n' %(i+1))
                final.write('|{0:^20s} | {1:^20s} | {2:^20s} | {3:^20s} | {4:^20s}%d\r\n'.format("Year", "Wage/Salary", "Cost of Vegetables", "Tuition and Fees", "Steer & Heifer Slaughter")% (i+1))
                for perc in range(len(common)):
                    final.write('|{0:^20s} | {1:^20.2f} | {2:^20.2f} | {3:^20.2f} | {4:^20.2f}%d\r\n'.format(common[perc], wage4[perc], veg4[perc], college4[perc], slag4[perc])% (i+1))
                final.close()
    except:
        print("Error")


filename = 'FINALME30_MAHMOOD_MEW_NGUYEN.txt'

sendit(filename)


commonint = []  

commoncvt(common, commonint)

def scalar(new, orig):
    for kilo in orig:
        new.append(kilo*10)
    
veg_graph = []
cow_graph = []

scalar(veg_graph, veg_annual)
scalar(cow_graph, slag_annual)       

uni_graph = tuit[0:-5]

x1 = np.array(commonint[0:-2])
y1 = np.array(li8[0:-4])
style.use('fivethirtyeight')

l_1=[1973,1974,1975,1976,1977,1978,1979,1980,1981,1982,1983,1984,1985,1986,1987,1988,1989,1990]
l_2=[5.61,6.95,6.45,5.65,7.36,8.04,8.26,9.15,5.22,4.65,5.55,4.09,2.88,6,4.69,3.81,4.42,3.59]
xs = np.array(l_1, dtype=np.float64)
ys = np.array(l_2, dtype=np.float64)

def best_fit_slope(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    b = mean(ys) - m*mean(xs)
    return m,b

m,b = best_fit_slope(xs,ys)
regression_line = [(m*x)+b for x in xs]

pl.scatter(xs,ys)
pl.plot(xs,regression_line)
pl.show()
    
def slope_and_intercept_for_wage_index(x1,y1):
    for i in x1, y1:
        m = ((sum(x1[i]*y1[i]) - sum(x1[i])*sum(y1[i])) / ((sum(x1[i])*sum(x1[i])) - sum(x1[i]*x1[i])))
        b = sum(y1[i]) - m*sum(x1[i])
    return m, b
m, b =slope_and_intercept_for_wage_index(x1,y1)
best_line = [(m*x)+b for x in x1]

a1= np.array(commonint[0:-2])
b1 = np.array(veg_graph)

def slope_and_intercept_for_vegetables(a1,b1):
    for i in a1, b1:
        n = (((sum(a1[i])*sum(b1[i])) - sum(a1[i]*b1[i])) / ((sum(a1[i])*sum(a1[i])) - sum(a1[i]*a1[i])))
        c = sum(b1[i]) - n*sum(a1[i])
    return n, c
n, c=slope_and_intercept_for_vegetables(a1,b1)
second_best_line=[(n*a)+c for a in a1]


live= np.array(commonint[0:-2])
stock = np.array(cow_graph)

def slope_and_intercept_for_steers(live,stock):
    for i in live, stock:
        o = (41*(sum(live[i])*sum(stock[i])) 
             - sum(live[i]*stock[i])) / (((sum(live[i])*sum(live[i])) - sum(live[i]*live[i])))
        d = sum(stock[i]) - o*sum(live[i])
    return o, d
o, d=slope_and_intercept_for_vegetables(live,stock)
third_best_line=[(o*a)+d for a in live]


UC = np.array(commonint[0:-2])
CSU = np.array(uni_graph)

def slope_and_intercept_for_college(UC,CSU):
    for i in UC, CSU:
        p = (((sum(UC[i])*sum(CSU[i])) - sum(UC[i]*CSU[i])) / ((sum(UC[i])*sum(UC[i])) - sum(UC[i]*UC[i])))
        e = sum(CSU[i]) - p*sum(UC[i])
    return p, e
p, e=slope_and_intercept_for_college(UC,CSU)
fourth_best_line=[(p*a)+e for a in UC]



pl.scatter(x1,y1)
pl.plot(x1, best_line, label='Average Wage Index')

pl.title('Wages in US versus Cost of Resources')
pl.xlabel('Years')
pl.ylabel('Amount')

pl.scatter(a1,b1)
pl.plot(a1, second_best_line, label='Fresh Vegetables(10x)')

pl.scatter(live, stock)
pl.plot(live, third_best_line, label='Steers & Heifers(10x)')

pl.scatter(UC, CSU)
pl.plot(UC, fourth_best_line, label='Tuition and Fees')
pl.grid(True)
pl.legend()
pl.savefig("US_ECON.svg")
pl.show()
```
